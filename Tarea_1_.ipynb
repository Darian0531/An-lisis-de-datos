1. Ley de los gases ideales


La ecuación de los gases ideales relaciona presión, el volumen, la cantidad de sustancia y la temperatura mediante la expresión:


image.png


image.png

Realice una busqueda para obtener el valor de la constante de los gases (R)en la unidades requeridas. Determina el valor de n del sistema.

```python
#Datos dados en el enunciado
P = 1.2 #presión (atm)
R = 0.0821 #constante de los gases ideales en unidades L*atm/(mol*K)
T = 273 #temperatura (K)
V = 0.7 #volumen (L)
#Cálculo del número de moles usando la ecuación de gas ideal
n = (P*V)/(R*T)
#Se imprime el resultado con cuatro decimales
print(f'Número de moles: {n: .4} mol')
```

4. Notas finales de clases

Escribe el código usando la estructura de control if, elif y else para determinar la nota de clase de acuerdo con la calificación numérica final
obtenida.



*   Menos de 6 es NA
*   De 6 a menos de 7.6, es S

*   De 7.6 a menos de 8.6, es B
*   De 8.6 a 10, es B





```python
# Pedimos al usuario una calificación y la convertimos a número decimal (float)
cal= float(input("Ingresa la calificación (0-10):"))
if cal>= 8.6 and cal <=10:
  #Si está entre 8.6 y 10 (incluye 10)
  print ('MB')
elif cal>= 7.6 and cal <= 8.5:
  # Si está entre 7.6 y menor que 8.6
  print('B')
elif cal>=6 and cal <=7.5:
  # Si está entre 6.0 y menor que 7.6
  print('S')
else:
  # Si es menor que 6 ( o si no cae en los rangos anteriores)
  print('NA')
```

5. Número de Reynolds

El número de Reynolds *(Re)* se utiliza para caracterizar el régimen de flujo (laminar, de transición o turbulento)

image.png

image.png

Calcule el valor del número de Reynolds.

```python
#Se definen la variables a usar
rho = 167
v = 40
D = 9.8
µ = 2.3

Re = rho*v*D/µ #Cálculo de número de Reynolds (adimensional)
#Imprime el resultado con tres decimales
print(f'El número de Reynolds es: {Re:.3f}')
```

6. Determinación del régimen de flujo

image.png

image.png

```python
#Se piden las propiedades del fluido y la tuberia
rho = float(input('Densidad del fluido en kg/m3: '))
v = float(input('Velocidad promedio del fluido en m/s: '))
D = float(input('Diámetro interno de la tubería en m: '))
mu = float(input('Viscosidad dinámica del fluido en Pa/s: '))

Re = rho*v*D/mu
#Clasificamos el tipo de flujo según Re
if Re < 2300:
  print('El flujo es  laminar')
elif Re >= 2300 and Re <= 4000:
  print('El flujo es de transición')
else:
  print('El flujo es  turbulento')
```
