```python
# 1
#Listas dadas
L1= [7.2, 7.8, 6.8, 8.0, None, '8.2', 5.6, 8.2, 7.7, 7.5, None, 5.8]
L2= ['6.8', None, 6.8, 6.1, 7.9, 9.4, None]
L3 = [8.5, '9.0', None, 7.7]

#Funcin para eliminar None usando un bucle for

def quitar_None(lista):
  #Elimina valores de None y convierte strings numricos a float
  lista_limpia = [] # Initialize an empty list
  #Se recorre cada elemento

  for elemento in lista:
    #Si el valor NO es None, lo agregamos
    if elemento is not None:
      lista_limpia.append(float(elemento)) #convertir a nmero
  return lista_limpia
#Aplicar funcin a cada lista
L1_limpia = quitar_None(L1)
L2_limpia = quitar_None(L2)
L3_limpia = quitar_None(L3)

#Mostrar resultados claramente
print('L1 limpia:', L1_limpia)
print('L2 limpia:', L2_limpia)
print('L3 limpia:', L3_limpia)
```

```python
# 2
# Juntar todos los valores en una sola lista
todos = []
for x in L1_limpia:
  todos.append(x)
for x in L2_limpia:
  todos.append(x)
for x in L3_limpia:
    todos.append(x)
print('Lista completa:', todos)
```

```python
#3
# Pedir al usuario los 9 valores de la matriz 3x3
matriz = []
print ('Ingrese los elementos de la matriz 3x3:')
#Leer filas con un bucle for
for i in range (3):
  fila=[] #Quiere decir fila vaca

  for j in range(3):
    valor= float(input(f'Elemento [{i+1},{j+1}]:'))
    fila.append(valor)
  matriz.append(fila)

#Mostrar la matriz
print('\nMatriz ingresada:')
for fila in matriz:
  print(fila)

  #Regla de Sarrus

  a= matriz[0][0]
  b= matriz[0][1]
  c= matriz[0][2]

  d= matriz[1][0]
  e= matriz[1][1]
  f= matriz[1][2]

  g= matriz[2][0]
  h= matriz[2][1]
  i= matriz[2][2]

  det= (a*e*i +b*f*g + c*d*h)- (c*e*g + b*d*i + a*f*h)
  print('\nDeterminante usando regla de Sarrus:', det)
```

```python
import numpy as np
#Convertir la matriz a arreglo NumPy
A= np.array(matriz)
#Calcular determinante con NumPy
det_numpy= np.linalg.det(A)

print('Determinante usando Numpy:', det_numpy)
```

```python
#5
# Estados de Mxico (con mayscula inicial)
estados = [
    "Aguascalientes","Baja California","Baja California Sur","Campeche","Chiapas","Chihuahua",
    "Coahuila","Colima","Durango","Guanajuato","Guerrero","Hidalgo","Jalisco","Estado de Mxico",
    "Michoacn","Morelos","Nayarit","Nuevo Len","Oaxaca","Puebla","Quertaro","Quintana Roo",
    "San Luis Potos","Sinaloa","Sonora","Tabasco","Tamaulipas","Tlaxcala","Veracruz","Yucatn",
    "Zacatecas","Ciudad de Mxico"
]

# Capitales ms comunes (tambin con mayscula)
capitales = [
    "Aguascalientes","Mexicali","La Paz","Campeche","Tuxtla Gutirrez","Chihuahua",
    "Saltillo","Colima","Durango","Guanajuato","Chilpancingo","Pachuca",
    "Guadalajara","Toluca","Morelia","Cuernavaca","Tepic","Monterrey",
    "Oaxaca","Puebla","Quertaro","Chetumal","San Luis Potos",
    "Culiacn","Hermosillo","Villahermosa","Ciudad Victoria","Tlaxcala",
    "Xalapa","Mrida","Zacatecas","CDMX"
]

# Pedir texto al usuario
texto = input("Escribe un estado o una capital de Mxico: ")

# Normalizar solo para comparar
texto_comp = texto.strip().lower()

# Banderas
es_estado = False
es_capital = False

# Buscar en estados
for e in estados:
    if texto_comp == e.lower():
        es_estado = True

# Buscar en capitales
for c in capitales:
    if texto_comp == c.lower():
        es_capital = True

# Resultado
if es_estado and es_capital:
    print(f"'{texto}' corresponde a un ESTADO y tambin a una CAPITAL.")
elif es_estado:
    print(f"'{texto}' corresponde a un ESTADO de Mxico.")
elif es_capital:
    print(f"'{texto}' corresponde a una CAPITAL de Mxico.")
else:
    print(f"'{texto}' NO corresponde a un estado ni a una capital.")
```

```python
#6
# Lista vaca para guardar los nmeros
numeros = []

# Variable para llevar la suma total
suma = 0

# Bucle while: se repite mientras la suma sea menor o igual a 100
while suma <= 100:

    # Pedir nmero al usuario
    num = float(input("Ingresa un nmero: "))

    # Agregar el nmero a la lista
    numeros.append(num)

    # Sumar al total
    suma += num

# Mostrar resultados finales
print("\nLista completa de nmeros:", numeros)
print("Suma total:", suma)
```

```python
# Crear lista vaca
lista = []

# Variable de control del men
opcion = ""

# Bucle while: se repite hasta que el usuario elija salir
while opcion != "4":

    # Mostrar men
    print("\nMen:")
    print("1. Agregar nmero")
    print("2. Eliminar ltimo nmero")
    print("3. Mostrar lista")
    print("4. Salir")

    # Pedir opcin al usuario
    opcion = input("Seleccione una opcin: ")

    # Opcin 1: Agregar nmero
    if opcion == "1":
        num = float(input("Ingrese un nmero: "))
        lista.append(num)
        print("Nmero agregado correctamente.")

    # Opcin 2: Eliminar ltimo nmero
    elif opcion == "2":
        if len(lista) > 0:
            eliminado = lista.pop()
            print("Se elimin el nmero:", eliminado)
        else:
            print("La lista est vaca, no se puede eliminar.")

    # Opcin 3: Mostrar lista
    elif opcion == "3":
        print("Contenido actual de la lista:", lista)

    # Opcin 4: Salir
    elif opcion == "4":
        print("Saliendo del programa...")

    # Opcin invlida
    else:
        print("Opcin no vlida. Intente de nuevo.")
```
